```python
                                    #### Accessing Huggingface LLMs using Langchain 
#Text To Text 
import os
from langchain_huggingface import HuggingFaceEndpoint
from langchain_core.messages import HumanMessage, SystemMessage
from langchain_huggingface import ChatHuggingFace


HuggingfaceHub_API_Token=os.getenv("HUGGINGFACEHUB_API_TOKEN")
if HuggingfaceHub_API_Token is None:
    raise ValueError("HUGGINGFACEHUB_API_TOKEN environment variable not set.")
else:
    os.environ["HUGGINGFACEHUB_API_TOKEN"] = HuggingfaceHub_API_Token;


#It is an LLM COnnector-Connects Code to a model hosted on HiggingFace Servers
llm=HuggingFaceEndpoint(
    repo_id="meta-llama/Llama-3.1-8B-Instruct",
    #repo_id="mistralai/Mistral-7B-v0.1", //This is Base Model which doesnt not take Instruction use "mistralai/Mistral-7B-Instruct-v0.3"
    # repo_id="deepseek-ai/DeepSeek-V3-0324",
    task="text-generation",
     max_new_tokens=512
)

# It is Wrapper on HuggingFaceEndpoint, which add Chat Capabilities.
chat_model=ChatHuggingFace(llm=llm) 

messages=[SystemMessage(content="You are a helpful assistant") , HumanMessage(content="What is the capital of India?")]

# messages=[SystemMessage(content="You are a helpful cook") , HumanMessage(content="What is the capital of India?")]   #Observe Output
ai_msg=chat_model.invoke(messages)
 
print(ai_msg.content)

#How it Works 
# Your Messages
#      ↓
# ChatHuggingFace        ← formats messages into prompt
#      ↓
# HuggingFaceEndpoint    ← sends to HuggingFace API
#      ↓
# HuggingFace Server     ← runs the model
#      ↓
# AIMessage              ← returned back to you
```

```python
                                        #### Accessing Groq  LLMs using Langchain
 # Text To Text
import os
from langchain_groq import ChatGroq
from langchain_core.messages import SystemMessage,HumanMessage

# print(os.getcwd()) # Check the current working directory

GROQ_API_KEY_Token=os.getenv("GROQ_API_KEY")
if GROQ_API_KEY_Token is None:
    raise ValueError("GROQ_API_KEY_Token environment variable not set.")
else:
    os.environ["GROQ_API_KEY"] = GROQ_API_KEY_Token;  

llm=ChatGroq(model="llama-3.1-8b-instant")

messages=[SystemMessage(content="You are a helpful Mathematician"),
          HumanMessage(content="What is Pythogonus Theorem? ")]

message=llm.invoke(messages)

print(message.content)
```

```python
                                                        #### Accessing OpenRouter  LLMs using Langchain
#Text To Text 
import os
from  langchain_openrouter import ChatOpenRouter
from langchain_core.messages import SystemMessage,HumanMessage   

OpenRouter_API_Key=os.getenv("OPENROUTER_API_KEY")
if OpenRouter_API_Key is None:
    raise ValueError("OPENROUTER_API_KEY environment variable not set.")
else :
    os.environ["OPENROUTER_API_KEY"] = OpenRouter_API_Key;

llm=ChatOpenRouter(model="tencent/hy3-preview")
messages=[SystemMessage(content="You are a Air Pilot") , HumanMessage(content="What are basics of flying?")]
ai_msg=llm.invoke(messages)
print(ai_msg.content)


```

```python
                                    #### Accessing Google Gemini LLMs using Langchain

#Text To Text
import os
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.messages import HumanMessage,SystemMessage

Google_API_Token_Key=os.getenv("GOOGLE_API_KEY")
if(Google_API_Token_Key is None):
    raise ValueError("GOOGLE_API_KEY environment variable not set.")
else:
    os.environ["Google_API_KEY"] = Google_API_Token_Key;

llm=ChatGoogleGenerativeAI(model="gemini-2.5-flash", max_output_tokens=2000)
messages=[SystemMessage(content="You are Prime Minister of India") , HumanMessage(content="Explain me the roles of PM.")]
ai_msg=llm.invoke(messages)
print(ai_msg.content)
```
