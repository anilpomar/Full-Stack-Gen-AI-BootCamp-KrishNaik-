```python
import pandas as pd
import numpy as np
import os
import gensim
import nltk
```

```python
#Download Resources from nltk(Natural Language ToolKit) Package used to Text Preprocessing
# (TokeniZation,word level Tokenization,Sentense level Tokenization ....)
nltk.download("all")
```

```python
from nltk import sent_tokenize #Splits text into sentences and no modification in oiriginal Text
from gensim.utils import simple_preprocess #Tokenizes text into words and lowercases text ,it removes punctuation
```

```python
for file in os.listdir('C:\Mine\AI\Full Stack Gen AI  BootCamp (KrishNaik)\Practicals\Assignment\Custom_Word2Vec_Training\Data'):
    print(file)
```

```python
#Now Load and Read data from both Data Files Together and perform tokenization
#Step 1: Load Both files into Memory
#Step 2: Convert paragraph into sentense using sent_tokenize from nltk library
#step 3: Convert each sentense into words(Tokens) using simple_process from gensim.utils library

folder_path = r"C:\Mine\AI\Full Stack Gen AI  BootCamp (KrishNaik)\Practicals\Assignment\Custom_Word2Vec_Training\Data"
story = []
for filename in os.listdir(folder_path):
  file_path=os.path.join(folder_path,filename)
  with open(file_path,encoding='unicode_escape') as f:
    corpus=f.read()
  raw_sent=sent_tokenize(corpus)
  for sent in raw_sent:
    story.append(simple_preprocess(sent))
    
    
```

```python
#Loads Model into memory now
#All Parameters those are mentioned inside model are Hyperparameters
#window : This parameter is hyper parameters which 
#min_count : Words Considered only whne more that mentioned number (ex: min_count=5)
        # sunny->100
        # Divyansh->6
        # Rahul->3
                    #Rahul is not considered since its frequency of occurance is 3
#Vector_Size: model will be trained only with that dimension (i,e., Features)

# Word2Vec(..., sg=0)  # Here sg=0 defines word2vec selected for training is CBOW #by default if mentioned nothing inside model parameter
# Word2Vec(..., sg=1)  # Here sg=0 defines word2vec selected for training is Skip-gram
custom_model=gensim.models.Word2Vec(
   window=10,
   min_count=5,# min_count=5 is default value
   vector_size=150
   #sg=1
)
```

```python
#Build Vocabulary
custom_model.build_vocab(story)
```

```python
#Lets Train Model now
#total_examples : This is mandate otherwise model wont be trained
#epocs: No of times model should train on data which is hyper parameter
custom_model.train(story, total_examples=custom_model.corpus_count, epochs=10)
```

```python
# What's Happening — The Problem
# All scores dropped at 30 epochs! This is a sign of instability due to small data.
# 5 epochs  → model underfit  (not enough learning)
# 30 epochs → model confused  (too many updates on tiny data)
# With only 2 story files, the model has very few sentences to learn from.
#  More epochs means it keeps re-adjusting the same word vectors back and forth, causing them to drift rather than converge.
```

```python
# There is Reduce in vector values As epochs increase. in Above 2 Cells.W
# What's Happening — Learning Rate Decay
# Word2Vec uses a decreasing learning rate during training:
# python# Default values in gensim
# alpha     = 0.025   # starting learning rate
# min_alpha = 0.0001  # minimum learning rate As epochs increase,the learning rate gradually decreases:
# Epoch 1  → alpha = 0.025      (large updates)
# Epoch 5  → alpha = 0.012      (medium updates)
# Epoch 10 → alpha = 0.005      (small updates)
# Epoch 50 → alpha = 0.0001     (tiny updates, near frozen)
# So at epoch 10, the model is making smaller adjustments to vectors than at epoch 5 → values shift less → appear slightly reduced/stabilized.
```
