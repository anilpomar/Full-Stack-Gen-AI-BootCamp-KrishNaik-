##Average word2vec 
##This Technique converts every sentense into vector.

#Steps to Perform
#1.First Split sentense into words by newlines or semicolons or spaces
#2.Create a vector for each word in the sentense.-->Then the average of all vectors to get a single vector for the sentense..

```python
#Convert Sentense to Vector
sentense ="This is Anil Working as Principal Software Engineer in Dell Technologies"

#Steps to Perform
#1.First Split sentense into words by newlines or semicolons or spaces
#2.Create a vector for each word in the sentense.-->Then the average of all vectors to get a single vector for the sentense..
```

```python
words=sentense.lower().split()
```

```python
word_Vectors=[]

for word in words:  
        print(word)
        word_Vectors.append(model[word])
```

```python
word_Vectors
```

```python
#Applying Average Concept
import numpy as np
sentence_vector =  np.mean(word_Vectors,axis=0) # axis=0 means Summamation on Column basaed and take average of it 
sentence_vector  #Now this vector representing Senctence vector
```
