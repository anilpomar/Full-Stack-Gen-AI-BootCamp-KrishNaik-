```python
from sklearn.preprocessing import OneHotEncoder
import numpy 
```

```python
documents = ["people watch movie", 
             "people watch cricket", 
             "people like movie", 
             "people like cricket"]
```

```python
#Convert documents into tokens
tokens=[sentence.split() for sentence in documents]
```

```python
tokens
```

```python
#Convert tokens into 2-D array Metrics
all_Words=[[word] for sentence in tokens for word in sentence]
```

```python
all_Words
```

```python
#OneHotEncorder Object
encoder=OneHotEncoder(sparse_output=False)
```

```python
encoder.fit(all_Words)
```

```python
#Check Vocabulary
# we can run categories with index 0 also it gives obejct of array
print ("Vocabulary: ", encoder.categories_[0])
```

```python
#Representing OneHotEncoder

for sentence in tokens:
    encoded_sentence = encoder.transform([[word] for word in sentence])
    print("\nSentence:", sentence)
    print("Encoded:", encoded_sentence)

```
