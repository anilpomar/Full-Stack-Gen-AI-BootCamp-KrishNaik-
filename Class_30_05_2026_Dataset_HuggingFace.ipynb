```python
shuffled = datasetForPreprocessing["train"].shuffle(seed=42)
```

```python
#Select subset of data from dataset
subset_train = datasetForPreprocessing["train"].shuffle().select(range(5000))
```

```python
print(subset_train)
```

```python
#Data from Dataset can be filtered with specific criteria
#Here we are fetching data rows that has text length less than 100
short_reviews = datasetForPreprocessing["train"].filter(lambda x: len(x["text"]) < 100)
```

```python
short_reviews
```

```python
#Here we are fetching data row that has Positive reviews (label =1 determines positivity) and Text length less than 200
#Breaking Logic
# x["label"] == 1 — Row has positive label (1)
# len(x["text"]) < 200 — Text length is less than 200 characters
# and — Both conditions must be true

short_positive = datasetForPreprocessing["train"].filter(lambda x: x["label"] == 1 and len(x["text"]) < 200)
```

```python
short_positive
```

```python
def add_word_count(example):
    example["word_count"] = len(example["text"].split())
    return example
```

```python
#Here it adds (add_wrod_count) column to dataset
dataset = datasetForPreprocessing.map(add_word_count)
```

```python
print(dataset)
```

```python
#Here we are reading no of words in First Row from Train Dataset
print(dataset["train"][0]["word_count"])
```
