```python
import pandas as pd
import numpy as np
import os
import gensim
import nltk
```

```python
#Download Resources from nltk(Natural Language ToolKit) Package used to Text Preprocessing
# (TokeniZation,word level Tokenization,Sentense level Tokenization ....)
nltk.download("all")
```

```python
from nltk import sent_tokenize #Splits text into sentences and no modification in oiriginal Text
from gensim.utils import simple_preprocess #Tokenizes text into words and lowercases text ,it removes punctuation
```

```python
for file in os.listdir('./Data/'):
    print(file)
```

```python
#Load Data Files into memory and read
f= open (r"C:\Mine\AI\Full Stack Gen AI  BootCamp (KrishNaik)\Practicals\Class-06-04-2026\Data\story1.txt",encoding='unicode_escape')
corpus=f.read()
corpus
```

```python
#Preprocess Data (Data split happend on fullstop(.))
rawsentense=sent_tokenize(corpus)
rawsentense
```

```python
#Convert each sentense into words(Tokens) using simple_preprocess from gensim.utils library
for words in rawsentense:
    print (simple_preprocess(words))
```

```python
#Now Load and Read data from both Data Files Together and perform tokenization
#Step 1: Load Both files into Memory
#Step 2: Convert paragraph into sentense using sent_tokenize from nltk library
#step 3: Convert each sentense into words(Tokens) using simple_process from gensim.utils library

folder_path = r"C:\Mine\AI\Full Stack Gen AI  BootCamp (KrishNaik)\Practicals\Class-06-04-2026\Data"
story = []
for filename in os.listdir(folder_path):
  file_path=os.path.join(folder_path,filename)
  with open(file_path,encoding='unicode_escape') as f:
    corpus=f.read()
  raw_sent=sent_tokenize(corpus)
  for sent in raw_sent:
    story.append(simple_preprocess(sent))
    
    
```

```python
#Loads Model into memory now
#All Parameters those are mentioned inside model are Hyperparameters
#window : This parameter is hyper parameters which 
#min_count : Words Considered only whne more that mentioned number (ex: min_count=5)
        # sunny->100
        # Divyansh->6
        # Rahul->3
                    #Rahul is not considered since its frequency of occurance is 3
#Vector_Size: model will be trained only with that dimension (i,e., Features)

# Word2Vec(..., sg=0)  # Here sg=0 defines word2vec selected for training is CBOW #by default if mentioned nothing inside model parameter
# Word2Vec(..., sg=1)  # Here sg=0 defines word2vec selected for training is Skip-gram
custom_model=gensim.models.Word2Vec(
   window=10,
   min_count=5,
   vector_size=150
   #sg=1
)
```

```python
#Build Vocabulary
custom_model.build_vocab(story)
```

```python
#Total Number of Sentences inside data file
custom_model.corpus_count
```

```python
#Lets Train Model now
#total_examples : This is mandate otherwise model wont be trained
#epocs: No of times model should train on data which is hyper parameter
custom_model.train(story, total_examples=custom_model.corpus_count, epochs=10)
```
