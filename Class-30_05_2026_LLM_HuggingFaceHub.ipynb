```python
from langchain_huggingface import ChatHuggingFace
from langchain_huggingface import HuggingFaceEndpoint
```

```python
llm =HuggingFaceEndpoint( repo_id="deepseek-ai/DeepSeek-R1",
    task="text-generation",
    max_new_tokens=512,
                          huggingfacehub_api_token=ReadToken
    )
```

```python
chatmodel=ChatHuggingFace(llm=llm)
```

```python
chatmodel.invoke("Explain Hugging Face in simple words")
```

```python
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough


template="""Question: {question}
give me a answer in detail manner and in step by step manner"""

prompt=PromptTemplate(template=template,input_variables=["question"])

llm_chain = (
        {"question": RunnablePassthrough()}
        | prompt
        | chatmodel
        | StrOutputParser()
    )

question="List out Top 5 IIT's in India"

llm_chain.invoke({"question":question})
```
